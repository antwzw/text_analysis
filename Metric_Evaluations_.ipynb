```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
data1 = pd.read_csv('/content/handcoded1.csv')
data1_cleaned = data1.dropna(subset=['Aspect'])
data1_cleaned = data1_cleaned.reset_index(drop = True)
data1_cleaned['Aspect'] = data1_cleaned['Aspect'].str.lower()
defined_aspects = data1_cleaned['Aspect'].unique()
defined_aspects
```

```python
data1_cleaned
```

```python
grouped_df1 = data1_cleaned.groupby('Review')['Aspect'].apply(', '.join).reset_index()
grouped_df1
```

```python
data2 = pd.read_csv('/content/pred_results.csv')
data2_cleaned = data2.dropna(subset=['Pred_Aspect'])
data2_cleaned = data2_cleaned.dropna(subset=['Summary'])
data2_cleaned['Pred_Aspect'] = data2_cleaned['Pred_Aspect'].str.lower()
data2_cleaned['Pred_Aspect'].unique() # the predicted aspects have more than what we defined and we need to correct those
```

```python
# I only used those Pred_Aspect that are in the defined aspect
# Function to check if any aspect in the list matches
def any_aspect_matches(predicted_aspects, defined_aspects):
    aspects = predicted_aspects.split(', ')
    return any(aspect in aspects for aspect in defined_aspects)

# Apply the function and filter the DataFrame
data2_filtered = data2_cleaned[data2_cleaned['Pred_Aspect'].apply(lambda x: any_aspect_matches(x, defined_aspects))]
data2_filtered = data2_filtered.reset_index(drop = True)
data2_filtered
```

```python
data2_filtered = data2_filtered.reset_index(drop = True)
grouped_df2 = data2_filtered.groupby('Review')['Pred_Aspect'].apply(', '.join).reset_index()
grouped_df3 = data2_filtered.groupby('Review')['Summary'].apply(', '.join).reset_index()
grouped_df2
```

```python
merged_predicted = grouped_df3.merge(grouped_df2, on='Review', how='inner')
merged_predicted
```

```python
merged_df = grouped_df1.merge(merged_predicted, on='Review', how='inner')
merged_df
```

```python
# Create a binary matrix for predicted labels
y_pred = np.zeros((len(merged_df), len(defined_aspects)))
for i, aspect in enumerate(defined_aspects):
    y_pred[:, i] = merged_df['Pred_Aspect'].apply(lambda x: int(aspect in x))
y_pred
```

```python
# Convert the aspect lists into binary multi-label representation
for Aspect in defined_aspects:
    merged_df[Aspect] = merged_df.apply(lambda row: int(Aspect in row['Aspect']), axis=1)

# Create the true and predicted labels as numpy arrays
y_true = np.zeros((len(merged_df), len(defined_aspects)))
for i, aspect in enumerate(defined_aspects):
    y_true[:, i] = merged_df['Aspect'].apply(lambda x: int(aspect in x))

# Create a binary matrix for predicted labels
y_pred = np.zeros((len(merged_df), len(defined_aspects)))
for i, aspect in enumerate(defined_aspects):
    y_pred[:, i] = merged_df['Pred_Aspect'].apply(lambda x: int(aspect in x))

# Calculate F1-scores for each aspect
f1_scores = []
for i, aspect in enumerate(defined_aspects):
    f1 = f1_score(y_true[:, i], y_pred[:, i])
    f1_scores.append((aspect, f1))

# Print F1-scores
for aspect, f1 in f1_scores:
    print(f"Aspect: {aspect}\tF1-score: {f1:.2f}")

# Print classification report
classification_rep = classification_report(y_true, y_pred, target_names=defined_aspects, zero_division=0)
print(classification_rep)
```
